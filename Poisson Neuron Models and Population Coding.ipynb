```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pickle
```

```python
filename = '/Users/nickwalker/Desktop/Coding/Python/Untitled Folder 1/data_w4.pickle'

with open(filename, 'rb') as f:
    data = pickle.load(f)
```

The data contains the output of 4 simulated neurons that emulate the behavior found in the cercal organs of a cricket
- each neuron was probed with a range of air velocity stimuli of uniform intensity and differing direction
- the data represents the firing rate of each of the neurons in response to each of the stimulus values
- each recording lasted 10 seconds and this process was repeated 100 times for each neuron-stimulus combination

```python
data
```

```python
pd.DataFrame(data["neuron1"]).head()
```

```python
avg1 = []
for i in range(0,24):
    val = data["neuron1"][:,i].mean()
    avg1.append(val)

avg2 = []
for i in range(0,24):
    val = data["neuron2"][:,i].mean()
    avg2.append(val)
    
avg3 = []
for i in range(0,24):
    val = data["neuron3"][:,i].mean()
    avg3.append(val)
    
avg4 = []
for i in range(0,24):
    val = data["neuron1"][:,i].mean()
    avg4.append(val)
```

```python
# plotting all 100 trials for a neuron:
for i in range(1, len(data["neuron4"][:,1])):
    plt.plot(data["stim"], data["neuron1"][i,:])
plt.show()
```

```python
plt.plot(data["stim"], avg1, '--k')
plt.plot(data["stim"], avg2, c='g')
plt.plot(data["stim"], avg3, c='r')
plt.plot(data["stim"], avg4, c='orange')
plt.legend(['neuron1','neuron2','neuron3', 'neuron4'])
plt.show()
```
