```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

<h4> Information and Entropy  </h4>
- formula for calculating entropy:



Below we'll calculate the amount of entropy of a Bernoulli Variable for different Probability Values:
- notice the entropy is maximized where probability = 1/2

```python
num_s = 6
prob_values = np.arange(0.01,1,0.01)
entropy_values = []

def entropy(p):
    return p*np.log2(p) + (1-p)*np.log2(1-p) # calculation for entropy

for p in prob_values:
    H = entropy(p)
    entropy_values.append(-H)

plt.plot(prob_values, entropy_values)
plt.xlabel('Probability')
plt.ylabel('Entropy')
plt.title('Bernoulli Entropy')
plt.plot()
```

<h4> Mutual Information </h4>
- quantifies how informative one variable is with respect to another variable
- used in predicting how spikes relate to sensory input from the outside world!



- where the term on the left is the entropy of the response, while the term on the right is the conditional response entropy

Below, we calculate the mutual information for a random variable 

```python
p_spike = 0.5 # probability of a spike
p_stim = 0.2 # probability of a stimulus
p_cond = np.arange(0.01, 1.0, 0.01) # Possible values for conditional distribution


H_spike = -entropy(p_spike) # entropy for a spike

MI = np.zeros(len(p_cond)) # initiate blank array

for i in range(len(p_cond)):
    p = p_cond[i] #iterate through probabilities
    H_spike_stim = -entropy(p) 
    H_spike_nostim = -entropy(p)
    MI[i] = H_spike - (p_stim*H_spike_stim + (1-p_stim)*H_spike_nostim)

plt.plot(p_cond, MI)
plt.title('Mutual Information')
plt.xlabel('Error Probability')
plt.ylabel('Mutual Information')
plt.plot()
```
